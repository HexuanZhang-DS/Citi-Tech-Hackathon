```python
def get_income(x):
    if x == '$8000 TO 9999':
        return 8000
    elif x == '$7000 TO 7999':
        return 7000
    elif x == '$25000 OR MORE':
        return 25000
    elif x == '$20000 - 24999':
        return 20000
    elif x == '$10000 - 14999':
        return 10000
    elif x == '$1000 TO 2999':
        return 1000
    elif x == '$3000 TO 3999':
        return 3000
    elif x == '$15000 - 19999':
        return 15000
    elif x == 'LT $1000':
        return 1
    elif x == '$4000 TO 4999':
        return 4000
    elif x == '$6000 TO 6999':
        return 6000
    else:
        return 5000       
```

```python
df["INCOME"] = df["INCOME"].apply(lambda x: get_income(x))
```

```python
df.head()
```

```python
correlations = df.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlations, annot=True)
```
